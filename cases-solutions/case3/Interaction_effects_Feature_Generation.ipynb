## Graphs for visualizing interaction effects (general names)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = {
    'Numerical Feature': [10, 15, 20, 25, 30, 10, 15, 20, 25, 30],
    'Categorical Feature': ['A', 'B', 'C', 'B', 'A','C', 'B', 'A', 'C', 'A'],
    'Target Value': [True, True, False, True, False, True, True, False, True, False]
}

# Create the DataFrame
df = pd.DataFrame(data)

# Calculate proportion of True (1) values for each combination of numerical and categorical features
grouped_data = df.groupby(['Numerical Feature', 'Categorical Feature'])['Target Value'].mean().reset_index()

# Plotting the line plot with error bars
plt.figure(figsize=(10, 6))
sns.lineplot(x='Numerical Feature', y='Target Value', hue='Categorical Feature', data=grouped_data, marker='o', ci='sd')

plt.xlabel('Numerical Feature')
plt.ylabel('Proportion of True Values (Target Value)')
plt.title('Interaction Effect: Numerical Feature vs. Target Value (Boolean) for different levels of Categorical Feature')
plt.legend(title='Categorical Level')
plt.grid(True)
plt.show()
```

In this plot:

The x-axis represents the "Numerical Feature."
The y-axis represents the proportion of True values (1s) in the "Target Value."
Each line represents a different level of the "Categorical Feature," and the markers (circles) indicate the proportion of True values for each level of the "Numerical Feature."
The error bars around the markers represent the variability in the proportions for each combination of "Numerical Feature" and "Categorical Feature."
By examining this plot, you can visually identify if there is an interaction effect between the categorical and numerical variables on the proportion of True values in the target variable. If the lines have different slopes or non-parallel trends, it suggests an interaction effect.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = {
    'Categorical Feature 1': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Categorical Feature 2': ['X', 'X', 'Y', 'Y', 'X', 'X', 'Y', 'Y'],
    'Target Value': [True, True, False, True, False, False, True, False]
}

# Create the DataFrame
df = pd.DataFrame(data)

# Calculate proportion of True (1) values for each combination of Categorical Feature 1 and Categorical Feature 2
proportions = df.groupby(['Categorical Feature 1', 'Categorical Feature 2'])['Target Value'].mean().reset_index()

# Plotting the bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x='Categorical Feature 1', y='Target Value', hue='Categorical Feature 2', data=proportions)

plt.xlabel('Categorical Feature 1')
plt.ylabel('Proportion of True Values (Target Value)')
plt.title('Interaction Effect: Categorical Feature 1 vs. Target Value for different levels of Categorical Feature 2')
plt.legend(title='Categorical Feature 2')
plt.grid(True)
plt.show()
```

In this plot:

The x-axis represents "Categorical Feature 1."
The y-axis represents the proportion of True values (1s) in the "Target Value."
The bars are grouped by different levels of "Categorical Feature 2."
This bar plot allows you to visualize the interaction effect between "Categorical Feature 1" and "Categorical Feature 2" on the proportion of True values in the target variable. If the heights of the bars vary significantly between different levels of "Categorical Feature 2" for each level of "Categorical Feature 1," it suggests an interaction effect.
