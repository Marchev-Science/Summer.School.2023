# 4.0 Functions & Stuff

Тези методи използват идеята, че в предисторията на самия процес и само на него се съдържа достатъчно информация за неговото прогнозиране. Това се дължи на свойството ‘инертност’, присъщо на всички реални системи. Икономическа инертност, психологическа, механична, електро-магнитне и пр.

Специфично е, че се налага някакво предварително правило върху относителното тегло в прогнозата на различните лагови стойности на променливата.

“Пълзяща средна” е метод за прогнозиране, където всяка точка на пълзящата средна на времевите интервали е средна аритметична (претеглена средна) на определен брой последователни стойности. Броят на наблюденията при изчисляването на пълзящата средна е определен така, че да минимизира ефекта на сезонността или на други смущения в интервала.

Експоненциалното изглаждане е гъвкав метод, където на наблюдаваните стойности се придават различни тегла при изчисляването на прогнозата. Този метод е близък до пълзящата средна, но позволява да се изгладят предишни неточности при прогнозирането. Има предимството да подобрява прогнозата, като към предишната прогноза добавя определена част от предишната прогностична грешка.

Всяка прогноза има свойството по подразбиране да е неточна или винаги да се различава от истинската стойност. С тази цел се изчислява средна относителна грешка по интерполация - R (периодите от обучаващата последователност) и екстраполация - S (периодите от контролната/тестова последователност).

За целта се използват функциите R и S със следното математическо описание - сборът на остатъците на прогнозата (Модул от разликате между прогнозата и реалната стойност) се дели на сбора на реалните стойности.:

```python
def wma(train
        ,ma = 3
        ,split = 0.2
        ,weights = [1/3,1/3,1/3]
       ):
    
    # функция за произведение на 2 листа с данни с цел създаване на претеглената среда
    # период 1 бива умножен по тегло 1, период 2 по тегло 2 и период 3 (най-скорошния) по тегло 3
    # функцията умножава всеки елемент от лист 1 по съотвестващия елемент от лист 2 (листовете трябва да са равни)
    
    def Sum_product(l1,l2):
        return sum([a * b for a, b in zip(l1, l2)])
    
    F = []
    X = train.values.tolist()
    for i in range(ma,(len(data))):

        # Y_hat е средната аритметична от пълзящите средни (като функция на итерирания ред и броя на средните) със теглата

        Y_hat = Sum_product(X[i-ma:i],weights)

        # Прогнозираната стойност бива присвоена към листа от прогнози

        F.append(Y_hat)

        if i >= len(train):

            # в списъка с познати стойности се присвоява прогнозата с цел екстраполация на вече екстраполирани данни

            X.append(Y_hat)
            
    return [np.nan] * ma + F
            
```

## Class voltron 

```python
class Regression_Full:
    
    def __init__(self, X, Y, split= 0.2):
        
        self.X = X
        self.Y = Y
        self.split = 0.2
        
        self.X_train = X[:-int(len(X)*split)]
        self.X_test  = X[-int(len(X)*split):]
        self.y_train = Y[:-int(len(X)*split)]
        self.y_test  = Y[-int(len(X)*split):]
        
        # линейна регресия
        
        self.LR = LinearRegression().fit(self.X_train.values.reshape(-1, 1), self.y_train.values.reshape(-1, 1))
        self.a = self.LR.intercept_[0] 
        self.b = self.LR.coef_[0][0]
        self.equation = f'{self.a} + {self.b} * x'
        self.linear_forecast = [self.a + self.b * value for value in list(X.values)]
        self.interpolation_error = mean_absolute_percentage_error(self.y_train, self.linear_forecast[:-int(len(X)*split)])
        self.extrapolation_error = mean_absolute_percentage_error(self.y_test, self.linear_forecast[-int(len(X)*split):])
        
        
        # логаритмична регресия
        
        self.LogLR = LinearRegression().fit(np.log(self.X_train.values.reshape(-1, 1)), self.y_train.values.reshape(-1, 1))
        self.log_a = self.LogLR.intercept_[0] 
        self.log_b = self.LogLR.coef_[0][0]
        self.log_lin_forecast = [self.log_a + self.log_b * np.log(value) for value in list(X.values)]
        
        self.log_interpolation_error = mean_absolute_percentage_error(self.y_train, self.log_lin_forecast[:-int(len(X)*split)])
        self.log_extrapolation_error = mean_absolute_percentage_error(self.y_test, self.log_lin_forecast[-int(len(X)*split):])
```

```python
our_regression = Regression_Full(X=data['X1 ( L ) '], Y=data['Y (GDP)'])

our_regression.log_interpolation_error, our_regression.log_extrapolation_error
```
